```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정 (윈도우 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv(r"C:\Users\water\OneDrive\바탕 화면\iris-classification\BP_data.csv")


# 히스토그램 그리기
plt.figure(figsize=(6, 4))
plt.hist(df['BMI'].dropna(), bins=30, color='skyblue', edgecolor='black')

plt.title('BMI 분포 (히스토그램)')
plt.xlabel('BMI')
plt.ylabel('빈도수')
plt.tight_layout()
plt.show()
```

```python
# 숫자형 칼럼만 추출해서 상관계수 계산
corr = df.corr(numeric_only=True)

# 히트맵 그리기
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)

plt.title(" 변수 간 상관관계 히트맵")
plt.tight_layout()
plt.show()
```

```python


# 2. 박스플롯 시각화
plt.figure(figsize=(6, 4))
sns.boxplot(x='Sex', y='Level_of_Stress', data=df)

# 3. 그래프 꾸미기
plt.title('성별에 따른 스트레스 수준 분포')
plt.xlabel('성별 (0=여성, 1=남성)')
plt.ylabel('스트레스 수준')
plt.tight_layout()

# 4. 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.stripplot(x='Smoking', y='BMI', data=df, jitter=True, alpha=0.6)

plt.title('흡연 여부에 따른 BMI 개별 분포 (점 그래프)')
plt.xlabel('흡연 여부 (0=비흡연, 1=흡연)')
plt.ylabel('BMI')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.stripplot(x='BMI', y='Smoking', data=df, jitter=0.25, alpha=0.6, orient='h')

plt.title('BMI에 따른 흡연 여부 분포 (점 그래프)')
plt.xlabel('BMI')
plt.ylabel('흡연 여부 (0=비흡연, 1=흡연)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='BMI', y='Level_of_Hemoglobin', data=df)

plt.title('BMI와 헤모글로빈 수치 간의 산점도')
plt.xlabel('BMI')
plt.ylabel('헤모글로빈 수치')
plt.tight_layout()
plt.show()
```
